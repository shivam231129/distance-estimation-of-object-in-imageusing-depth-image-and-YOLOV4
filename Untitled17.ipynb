```python
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_squared_error
def loss (clf,X_train,X_cv,y_train,y_cv):
    y_pred=clf.predict(X_cv)
    print("rmse cv", np.sqrt(mean_squared_error(y_cv, y_pred)))
    y_pred=clf.predict(X_train)
    print("rmse train", np.sqrt(mean_squared_error(y_train, y_pred)))
    
    y_pred=clf.predict(X_cv)
    print("rmsle test",np.sqrt(mean_squared_log_error(y_cv, y_pred)))
    y_pred=clf.predict(X_train)
    print("rmsle train",np.sqrt(mean_squared_log_error(y_train, y_pred)))
```
