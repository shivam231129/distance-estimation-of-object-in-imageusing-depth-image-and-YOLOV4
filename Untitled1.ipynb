```python
def my_func(s, z):
  total = 0
  len_str = len(s)

  for x, y in z:
    if x == 0:
      total += y
    else:
      total -= y
  total = total % len_str
  print(s[total:] + s[:total])


s = 'cutshort'
z = [[0, 3], [1, 11]]

if __name__ == "__main__":
  my_func(s, z)
```

```python
 def oddsums(n):
  total=0
  result=[]
  for i in range(1,n+1):
    odd= 2*i-1
    total=total+odd
    result.append(total)
  return result
```
